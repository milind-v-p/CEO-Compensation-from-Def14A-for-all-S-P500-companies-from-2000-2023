```python
from sec_api import QueryApi, ExtractorApi
import requests
from bs4 import BeautifulSoup
import re

# Initialize the QueryApi and ExtractorApi with your API key
api_key = "4c760d9988d08a39c264deca0af915ca06b5dac43aa0de77b1ea137d839ed88c"
query_api = QueryApi(api_key=api_key)
extractor_api = ExtractorApi(api_key=api_key)

# Define the query using the ticker symbol with date range from 1994 to 2023
query = {
    "query": "ticker:AAPL AND (formType:\"10-K\" OR formType:\"10-Q\") AND filedAt:[2023-01-01 TO 2023-12-31]",
    "from": "0",
    "size": "100",
    "sort": [{"filedAt": {"order": "desc"}}]
}

# Fetch the filings
filings = query_api.get_filings(query)

# List to store extracted words
extracted_words = []

# Function to fetch and parse text from URL
def fetch_text_from_url(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    return soup.get_text(separator=' ').split()

# Iterate through the filings and extract item 11 (Executive Compensation)
for filing in filings['filings']:
    if filing['formType'] == '10-K':
        url = filing['linkToHtml']  # or use 'linkToTxt' if plain text link is preferred
        
        try:
            # Fetch the section from the filing
            item_11_text = extractor_api.get_section(url, "11", "text")
            
            # Convert the text into an array of words
            item_11_words = item_11_text.split()  # Split by whitespace

            # Check if the section mentions a reference to a Proxy Statement
            if "Proxy" in item_11_words and "Statement" in item_11_words:
                # Parse the HTML to find the hyperlink to DEF 14A or relevant keywords
                response = requests.get(url)
                soup = BeautifulSoup(response.text, 'html.parser')
                
                # Find all <a> tags that link to DEF 14A or contain keywords related to executive compensation
                proxy_links = []
                for a_tag in soup.find_all('a', href=True):
                    if re.search(r'DEF\s*14A|Executive\s*Compensation', a_tag.text, re.IGNORECASE):
                        proxy_links.append(a_tag['href'])
                
                # Fetch and parse each referenced document
                for proxy_link in proxy_links:
                    if not proxy_link.startswith("http"):
                        proxy_link = f"https://www.sec.gov{proxy_link}"
                    referenced_text = fetch_text_from_url(proxy_link)
                    extracted_words.extend(referenced_text)
            else:
                # If no reference is found, add words to the extracted words list
                extracted_words.extend(item_11_words)
            
        except Exception as e:
            print(f"Error fetching item 11 (Executive Compensation) from URL {url}: {e}")

# Print the extracted words array
print(extracted_words)
```

```python
from sec_api import QueryApi, ExtractorApi
import requests
from bs4 import BeautifulSoup
import re

# Initialize the QueryApi and ExtractorApi with your API key
api_key = "4c760d9988d08a39c264deca0af915ca06b5dac43aa0de77b1ea137d839ed88c"
query_api = QueryApi(api_key=api_key)
extractor_api = ExtractorApi(api_key=api_key)

# Define the query using the ticker symbol with date range from 1994 to 2023
query = {
    "query": "ticker:AAPL AND (formType:\"10-K\" OR formType:\"10-Q\" OR formType:\"14A\") AND filedAt:[2023-01-01 TO 2023-12-31]",
    "from": "0",
    "size": "100",
    "sort": [{"filedAt": {"order": "desc"}}]
}

# Fetch the filings
filings = query_api.get_filings(query)

# List to store extracted words
extracted_words = []

# Function to fetch and parse text from URL
def fetch_text_from_url(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    return soup.get_text(separator=' ').split()

# Function to find and fetch text from DEF 14A filings
def fetch_def14a_text(ticker, year):
    def14a_query = {
        "query": f"ticker:{ticker} AND formType:\"DEF 14A\" AND filedAt:[{year}-01-01 TO {year}-12-31]",
        "from": "0",
        "size": "10",
        "sort": [{"filedAt": {"order": "desc"}}]
    }
    
    def14a_filings = query_api.get_filings(def14a_query)
    
    def14a_text = []
    for def14a_filing in def14a_filings['filings']:
        def14a_url = def14a_filing['linkToHtml']
        def14a_text.extend(fetch_text_from_url(def14a_url))
        
    return def14a_text

# Iterate through the filings and extract item 11 (Executive Compensation)
for filing in filings['filings']:
    if filing['formType'] == '10-K':
        url = filing['linkToHtml']  # or use 'linkToTxt' if plain text link is preferred
        
        try:
            # Fetch the section from the filing
            item_11_text = extractor_api.get_section(url, "11", "text")
            
            # Convert the text into an array of words
            item_11_words = item_11_text.split()  # Split by whitespace

            # Check if the section mentions a reference to a Proxy Statement
            if "Proxy" in item_11_words or "Statement" in item_11_words:
                # Extract the year from the filing date
                year = filing['filedAt'][:4]
                
                # Fetch text from DEF 14A filings for the same year
                def14a_text = fetch_def14a_text("AAPL", year)
                extracted_words.extend(def14a_text)
            else:
                # If no reference is found, add words to the extracted words list
                extracted_words.extend(item_11_words)
            
        except Exception as e:
            print(f"Error fetching item 11 (Executive Compensation) from URL {url}: {e}")

# Print the extracted words array
print(extracted_words)
```

```python
import requests
import pandas as pd

# Initialize the API key
api_key = "4c760d9988d08a39c264deca0af915ca06b5dac43aa0de77b1ea137d839ed88c"

# Function to fetch compensation data
def fetch_compensation_data(api_key, query):
    url = "https://api.sec-api.io/compensation/query"
    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {api_key}"
    }
    
    response = requests.post(url, headers=headers, json=query)
    
    if response.status_code == 200:
        return response.json()
    else:
        raise Exception(f"API error: {response.status_code} - {response.text}")

# Initialize empty list to store data
all_data = []

# Define initial query
query = {
    "query": "ticker:AAPL AND year:[1994 TO 2023]",
    "from": "0",
    "size": "50",
    "sort": [{"year": {"order": "desc"}}]
}

# Pagination handling
while True:
    compensation_data = fetch_compensation_data(api_key, query)
    
    if not compensation_data.get('executiveCompensations'):
        break
    
    all_data.extend(compensation_data['executiveCompensations'])
    
    # Increment 'from' parameter to fetch next set of data
    query['from'] = str(int(query['from']) + int(query['size']))

# Extract relevant information and create a DataFrame
data = []
for item in all_data:
    year = item.get('year')
    name = item.get('name')
    position = item.get('position')
    salary = item.get('salary')
    bonus = item.get('bonus')
    stock_awards = item.get('stockAwards')
    option_awards = item.get('optionAwards')
    non_equity_incentive_plan_comp = item.get('nonEquityIncentivePlanCompensation')
    change_in_pension_value = item.get('changeInPensionValueAndNonqualifiedDeferredCompensationEarnings')
    all_other_comp = item.get('allOtherCompensation')
    total = item.get('total')
    
    data.append([
        year, name, position, salary, bonus, stock_awards, option_awards, 
        non_equity_incentive_plan_comp, change_in_pension_value, all_other_comp, total
    ])

# Create a DataFrame
columns = [
    "Year", "Name", "Position", "Salary", "Bonus", "Stock Awards", "Option Awards", 
    "Non-Equity Incentive Plan Comp", "Change in Pension Value", "All Other Comp", "Total"
]
df = pd.DataFrame(data, columns=columns)

# Display the DataFrame
print(df)

# # Save the DataFrame to a CSV file
# df.to_csv("apple_executive_compensation_1994_2023.csv", index=False)
```
